```python
# Let's create a function for difference between max/min vehicle age.

def dif_age(age):
    return age.max() - age.min()
```

```python
# Apply dif_age function to df with groupby method.
df.groupby("Type")[["Vehicle_Age"]].agg(dif_age)
```
