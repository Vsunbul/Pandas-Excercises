>> Find the months that models that Toyota sold more than 500 in a month.

```python
data[(data.Quantity > 500) & (data.Make == "Toyota ")].head()
```

```python
data[(data.Quantity > 500) & (data.Make == "Toyota")].head(9)
```

>> Find the months that Toyota Rav4 sold more than 500 in a month.

```python
data[(data.Quantity > 500) & (data.Model == "Toyota Rav4")].Month.reindex
```

>> Find the highest percentage sales amount by year.

```python
data.groupby(["Make", "Year"]).agg({"Quantity":"sum", "Pct":"max"})
```

>> Find total sales of Toyota between the years 2010-2015.

```python
data.query("2010 < Year < 2015 & Make == 'Toyota'")
```

```python
data.query("2010 < Year < 2015 & Make == 'Toyota'").Quantity.sum()
```

>> Find the total sales in 2011

```python
data.query("Year == 2011").Quantity.sum()
```
